```python
# Step 3: Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
for col in df.columns:
    # For numerical columns (float64 and int64)
    if df[col].dtype == 'float64' or df[col].dtype == 'int64':
        df[col].fillna(df[col].mean(), inplace=True)
        
    # For categorical columns (object)
    elif df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)

# Check again for missing values after filling
missing_values_after_fill = df.isnull().sum()
print("\nMissing Values after filling:")
print(missing_values_after_fill)
```

```python
# Drop columns with a high number of missing values
df.drop(columns=['other_fuel3', 'wepp_id', 'generation_gwh_2013', 'generation_gwh_2019', 'estimated_generation_gwh'], inplace=True)
```

```python
from scipy.stats import skew

numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
skewness = df[numerical_cols].apply(lambda x: skew(x))

# Print the skewness for each column
print("Skewness:")
print(skewness)
```

```python
#Handle skewness 
# Example: Log transformation for columns with skewness > 1
skewed_cols = skewness[abs(skewness) > 1].index
df[skewed_cols] = np.log1p(df[skewed_cols])
# Print skewness after transformation
skewness_after = df[numerical_cols].apply(lambda x: skew(x))
print("\nSkewness after transformation:")
print(skewness_after)
```

```python
from scipy.stats import zscore
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
z_scores = zscore(df[numerical_cols])

# Find indices of outliers
outlier_indices = (abs(z_scores) > 3).any(axis=1)

# Print the indices of rows with outliers
print("Outlier Indices:")
print(df.index[outlier_indices].tolist())
```

```python
threshold = 3
df[numerical_cols] = df[numerical_cols].clip(lower=-threshold, upper=threshold)

# Print confirmation message
print("\nOutliers handled.")
```

```python
#check multicollinearity
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X_regression_standardized_df = pd.DataFrame(X_regression_standardized, columns=X_regression.columns)

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data["feature"] = X_regression_standardized_df.columns
vif_data["VIF"] = [variance_inflation_factor(X_regression_standardized_df.values, i) for i in range(X_regression_standardized_df.shape[1])]

# Print VIF values
print("VIF Values:")
print(vif_data)
```

```python
# It appears that there isn't significant multicollinearity present in your dataset. This is indicated by the fact that none of the VIF values are substantially greater than 10.
```

```python
# Identify features with high VIF
high_vif_features = vif_data[vif_data["VIF"] > 10]["feature"]

# Remove one of the correlated variables
X_regression_no_multicollinearity = X_regression_standardized_df.drop(columns=high_vif_features)

# Print confirmation message
print(f"The following features were removed due to high VIF: {list(high_vif_features)}")

# Recheck VIF after removing the variable
vif_data_no_multicollinearity = pd.DataFrame()
vif_data_no_multicollinearity["feature"] = X_regression_no_multicollinearity.columns
vif_data_no_multicollinearity["VIF"] = [variance_inflation_factor(X_regression_no_multicollinearity.values, i) for i in range(X_regression_no_multicollinearity.shape[1])]
print("\nVIF Values after handling multicollinearity:")
print(vif_data_no_multicollinearity)
```
