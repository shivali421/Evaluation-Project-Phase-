```python
# Define numerical and categorical columns
numerical_cols = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 'WindGustSpeed',
                  'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm', 'Pressure9am',
                  'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm']
categorical_cols = ['Date','Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday', 'RainTomorrow']

# Fill missing values in numerical columns with the mean of the column
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].mean())

# Fill missing values in categorical columns with the most frequent value
df[categorical_cols] = df[categorical_cols].apply(lambda x: x.fillna(x.value_counts().index[0]))

# Check if there are any remaining missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# One-Hot Encoding for Location column
df = pd.get_dummies(df, columns=['Date','Location'], drop_first=True)
```

```python
# Label Encoding for WindGustDir, WindDir9am, WindDir3pm columns
wind_directions = ['WindGustDir', 'WindDir9am', 'WindDir3pm']
for col in wind_directions:
    df[col] = df[col].astype('category').cat.codes
```

```python
# Binary Encoding for RainToday and RainTomorrow columns
binary_encoding = {'No': 0, 'Yes': 1}
df['RainToday'] = df['RainToday'].map(binary_encoding)
df['RainTomorrow'] = df['RainTomorrow'].map(binary_encoding)
```

```python
from scipy.stats import skew
# Calculate skewness for each numerical column
skewness = df[numerical_cols].apply(lambda x: skew(x.dropna()))
```

```python
# Filter columns with skewness greater than 0.5
skewed_cols = skewness[abs(skewness) > 0.5].index

# Apply log transformation to handle skewness
df[skewed_cols] = np.log1p(df[skewed_cols])

# Check the skewness again after transformation
skewness_after = df[numerical_cols].apply(lambda x: skew(x.dropna()))
```

```python
# Define a function to handle outliers using IQR method
def handle_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = np.where((df[column] < lower_bound) | (df[column] > upper_bound), df[column].median(), df[column])
```

```python
# Apply the handle_outliers function to each numerical column
for col in numerical_cols:
    handle_outliers(col)
```

```python
# Check if outliers have been handled
outliers_handled = df[numerical_cols].apply(lambda x: (x < x.quantile(0.25) - 1.5 * (x.quantile(0.75) - x.quantile(0.25))) |
                                                    (x > x.quantile(0.75) + 1.5 * (x.quantile(0.75) - x.quantile(0.25))))
```

```python
# Print columns with outliers after handling
print(outliers_handled.any())
```

```python
from scipy.stats.mstats import winsorize

# Define the columns with outliers
columns_with_outliers = ['MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 'WindSpeed9am', 
                         'Humidity9am', 'Pressure9am', 'Pressure3pm', 'Temp3pm']

# Apply winsorizing to cap extreme values
for col in columns_with_outliers:
    df[col] = winsorize(df[col], limits=(0.01, 0.01))

# Check if outliers have been handled
outliers_handled = df[columns_with_outliers].apply(lambda x: (x < x.quantile(0.01)) | (x > x.quantile(0.99)))

# Print columns with outliers after handling
print(outliers_handled.any())
```
