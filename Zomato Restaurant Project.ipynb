```python
# Define a function to handle skewness using log transformation
def handle_skewness_log(column):
    if column.skew() > 0.5:  # You can adjust this threshold based on your data
        column = np.log1p(column)
    return column

# List of numerical columns
numerical_columns = ['Restaurant ID', 'Country Code', 'Longitude', 'Latitude', 
                     'Average Cost for two', 'Price range', 'Aggregate rating', 'Votes']

# Apply skewness handling to numerical columns
for col in numerical_columns:
    merged_df[col] = handle_skewness_log(merged_df[col])
```

```python
#check outliers
def handle_outliers_log(column):
    Q1 = np.percentile(column, 25)
    Q3 = np.percentile(column, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (column < lower_bound) | (column > upper_bound)
    if any(outliers):
        column[outliers] = np.log1p(column[outliers])
    
    return column

# List of numerical columns
numerical_columns = ['Restaurant ID', 'Country Code', 'Longitude', 'Latitude', 
                     'Average Cost for two', 'Price range', 'Aggregate rating', 'Votes']

# Apply outlier handling to numerical columns
for col in numerical_columns:
    merged_df[col] = handle_outliers_log(merged_df[col])
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Calculate VIF for each feature
def calculate_vif(data):
    vif_data = data.copy()
    vif_data = vif_data.select_dtypes(include=['float64', 'int64'])
    vif_data = vif_data.dropna()

    vif_data['Intercept'] = 1
    vif = pd.DataFrame()
    vif["Variable"] = vif_data.columns
    vif["VIF"] = [variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])]
    return vif

vif_results = calculate_vif(features)

# Print the VIF results
print(vif_results)
```
